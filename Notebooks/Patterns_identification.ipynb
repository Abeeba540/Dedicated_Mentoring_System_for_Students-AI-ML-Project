```python
df = pd.read_csv('students.csv')
```

```python
df.head()
```

```python
# Pattern 1 - High Stress + Low Productivity

pattern_1 = df[df["stress_level"] >= 8 & (df["productivity_score"] <= 4)]
```

```python
print(pattern_1)
```

```python
len(pattern_1)
```

```python
# Pattern_2 - Low GPA + High Engagement

pattern_2 = df[(df["gpa"] <= 6) & (df["engagement_score"] >= 65)]

pattern_2
```

```python
len(pattern_2)
```

```python
# Pattern 3 - Strong Academics + Unclear Career Goals

pattern_3 = df[(df["gpa"] >=7.5) &(df["career_clarity"] <= 5)]

pattern_3
```

```python
len(pattern_3)
```

```python
pattern_summary = pd.DataFrame({
    "Pattern": [
        "High Stress + Low Productivity",
        "Low GPA + High Engagement",
        "Strong Academics + Unclear Career"
    ],
    "Number of Students": [
        len(pattern_1),
        len(pattern_2),
        len(pattern_3)
    ]
})

pattern_summary
```
