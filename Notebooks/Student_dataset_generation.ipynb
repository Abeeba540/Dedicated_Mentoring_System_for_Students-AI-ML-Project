```python
# Import necessary Libraries

import pandas as pd
import numpy as np
```

```python
# Set random number and define variable

np.random.seed(42)
n = 50
```

```python
# Randomly assigning programs

programs = ["B.Tech", "B.Sc", "MBA", "BCA", "MCA"]
```

```python
# Data Generation

df = pd.DataFrame({
    "student_id": [f"S{str(i).zfill(3)}" for i in range(1, n + 1)],
    "age": np.random.randint(18, 26, n),
    "program": np.random.choice(programs, n),
    "semester": np.random.randint(1, 9, n),

    # Academic
    "gpa": np.round(np.clip(np.random.normal(7.0, 1.2, n), 0, 10), 1),
    "attendance": np.clip(np.random.randint(50, 101, n), 0, 100),
    "assignments_completion": np.clip(np.random.randint(50, 101, n), 0, 100),

    # Wellness
    "stress_level": np.clip(np.random.randint(1, 11, n), 1, 10),
    "sleep_hours": np.round(np.clip(np.random.uniform(4, 9, n), 0, 10), 1),
    "mental_wellbeing": np.clip(np.random.randint(1, 11, n), 1, 10),

    # Productivity
    "productivity_score": np.clip(np.random.randint(1, 11, n), 1, 10),
    "distractions": np.clip(np.random.randint(1, 11, n), 1, 10),

    # Career
    "career_clarity": np.clip(np.random.randint(1, 11, n), 1, 10),
    "skill_readiness": np.clip(np.random.randint(1, 11, n), 1, 10),

    # Engagement
    "engagement_score": np.clip(np.random.randint(0, 101, n), 0, 100),
})

```

```python
# Save dataset

df.to_csv("students.csv", index=False)

```
